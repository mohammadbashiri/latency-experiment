```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('../Measurements/frame_checking.csv')
data[:5000].plot()
```

```python
dd = data.Chan1.values[3580:4100]*5/1024

fig, ax = plt.subplots(figsize=(10,5), dpi=500)
hfont = {'fontname':'Calibri'}
plt.xticks([0, 87, 175, 263, 350, 437, 525], [0, .5, 1, 1.5, 2, 2.5, 3], fontsize=20)
plt.yticks(fontsize=20)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.xlabel('Time (ms)', fontsize=20)
plt.ylabel('Signal (V)', fontsize=20)
plt.plot(dd, c='k', linewidth=3)
plt.tight_layout()

plt.savefig('per_frame.png')
```
